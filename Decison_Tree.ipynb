```python
#Build decision tree model to predict survival based on certain parameters
```

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
```

```python
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Age = LabelEncoder()
le_Pclass = LabelEncoder()
le_Embarked = LabelEncoder()
```

```python
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Pclass_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['Embarked_n'] = le_Embarked.fit_transform(inputs['Embarked'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Age','Pclass','Cabin','Name','PassengerId','Sex','SibSp','Parch','Ticket','Fare','Embarked'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```
